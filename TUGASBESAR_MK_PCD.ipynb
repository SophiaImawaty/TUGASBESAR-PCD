```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca citra dalam mode grayscale
image = cv2.imread('/content/x-ray-2.jpg', cv2.IMREAD_GRAYSCALE)

# Membuat daftar citra dan judul
images = [image]
titles = ['Citra Asli']

# 1. Penyesuaian Intensitas (Intensity Adjustment)
adjusted_image = cv2.convertScaleAbs(image, alpha=1, beta=40)
images.append(adjusted_image)
titles.append('Penyesuaian Intensitas')

# 2. Ekualisasi Histogram (Histogram Equalization)
equalized_image = cv2.equalizeHist(adjusted_image)
images.append(equalized_image)
titles.append('Ekualisasi Histogram')

# 3. Thresholding
_, thresholded_image = cv2.threshold(equalized_image, 127, 255, cv2.THRESH_BINARY)
images.append(thresholded_image)
titles.append('Thresholding')

# 4. Deteksi Tepi menggunakan Canny
edges = cv2.Canny(thresholded_image, 100, 200)
images.append(edges)
titles.append('Deteksi Tepi')

# Menampilkan citra hasil dalam satu figure
plt.figure(figsize=(25, 5))
for i in range(len(images)):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
